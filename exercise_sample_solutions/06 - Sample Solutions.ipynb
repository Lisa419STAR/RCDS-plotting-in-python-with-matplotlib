# Programming Experience

```python
import matplotlib.pyplot as plt

# Create the data for the languages
languages = ["Python", "R", "Matlab", "C++", "C", "Java", "Fortran"]
language_frequency = [152, 89, 66, 47, 30, 22, 15]

# Create the data for the experience level
expertise = ["Novice", "Intermediate", "Advanced"]
expertise_frequency = [108, 188, 125]

# Create the subplots
# Specify one row and two columns
# Use sharey to cause plots in the same row to share a y-axis
# Set the figure to be longer than it is high so it's not squashed
fig, axs = plt.subplots(1, 2, sharey='row', figsize= (10, 5))

# Set the y-label for the left figure
# The y-axis only needs to be specified for the left-most figure as this has the scale for both subplots
axs[0].set_ylabel("Number of Students")
# Set the x_label for and create the left-most bar chart
axs[0].set_xlabel("Most Used Language")
axs[0].bar(languages, language_frequency)

# Create the right-most bar chart
axs[1].bar(expertise, expertise_frequency)
axs[1].set_xlabel("Experience Level")
#this is 1d, dont make 2d by doing 0,0 oh
# Set the title for the figure
fig.suptitle("Survey Results")

# Edit the spacing such that the figures are adequately spaced
fig.tight_layout()
```
